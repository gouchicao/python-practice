# matplotlib & pandas 使用鸢尾花绘图

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

## 加载鸢尾花数据集

```python
# 加载数据 花萼长度、花萼宽度、花瓣长度、花瓣宽度
data = load_iris().data
data[:10]
```

```python
# 加载标签 山鸢尾（Setosa）、变色鸢尾（Versicolour）还是维吉尼亚鸢尾（Virginica）
label = load_iris().target
label
```

## 拆线图

```python
plt.figure()
plt.plot(data)
plt.show()
```

## 散点图

```python
# 只分析前两个特征
X = data[:, :2]
y = label
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
# s表示绘制的符号大小
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.get_cmap('cool'), 
            s=np.random.uniform(50, 200, len(X)), marker='*')
plt.show()
```

## 柱状图

```python
X = data[:20, :1]
plt.figure()
plt.bar(range(len(X)), X[:, 0])
plt.show()
```

* 每个柱不同颜色

```python
plt.figure()
for i in range(len(X)):
    plt.bar(i, X[i, 0])
plt.show()
```

* ylim

```python
plt.figure()
for i in range(len(X)):
    plt.bar(i, X[i, 0])
plt.ylim([4, 6])
plt.show()
```

* xticks

```python
plt.figure()
for i in range(len(X)):
    plt.bar(i, X[i, 0])
plt.ylim([4, 6])
plt.xticks(np.arange(0, len(X), step=5), rotation=90)
plt.show()
```

* text

```python
plt.figure()
for i in range(len(X)):
    plt.bar(i, X[i, 0])
    # 第2个参数是绘制文本的高度，第3个参数是要绘制的文本。
    plt.text(i, X[i, 0]+0.05, X[i, 0], ha='center')
plt.show()
```

```python
plt.figure()
for i in range(len(X)):
    plt.bar(i, X[i, 0], width=0.1)
    plt.text(i, X[i, 0]+0.05, X[i, 0], ha='center')
plt.plot(X[:, 0]-1, c='r', ls='--', lw=1, marker='.')
plt.show()
```

## 子图

* 拆线图

```python
plt.figure(figsize=(20, 10))
for i in range(data.shape[1]):
    plt.subplot(2, 2, i+1) # 参数分别是子图行数、子图列数、当前子图
    plt.plot(data[:, i])
plt.show()
```

* 散点图

```python
plt.figure(figsize=(20, 10))
markers = ['^', '*', 'o', '.']
for i in range(data.shape[1]):
    plt.subplot(2, 2, i+1)
    plt.scatter(range(len(data)), data[:, i], s=100, marker=markers[i])
plt.show()
```

* 移除轴的刻度

```python
plt.figure()
for i in range(data.shape[1]):
    plt.subplot(2, 2, i+1)
    plt.plot(data[:, i])
    
    if i==0 or i==1:
        # 移除X轴的刻度
        plt.xticks([])
    else:
        # 移除Y轴的刻度
        plt.yticks([])
plt.show()
```

* 关闭轴的刻度和边框

```python
plt.figure()
for i in range(data.shape[1]):
    plt.subplot(2, 2, i+1)
    plt.plot(data[:, i])
    # 关闭X和Y轴的刻度和边框
    plt.axis('off')
plt.show()
```

* 设置中文title

```python
import matplotlib
matplotlib.rcParams['font.sans-serif'] = 'Times-Roman'

plt.figure(figsize=(10, 6))
for i in range(data.shape[1]):
    plt.subplot(2, 2, i+1)
    plt.plot(data[:, i])
    plt.title('子图{}'.format(i+1))
plt.show()
```

* 分析特征相关性

```python
plt.figure(figsize=(20, 10))
for r in range(data.shape[1]):
    for c in range(data.shape[1]):
        plt.subplot(data.shape[1], data.shape[1], r*data.shape[1]+c+1)
        plt.scatter(data[:, r], data[:, c], c=label, s=50)
plt.show()
```

## 分类算法

* 拟合、预测

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(solver='lbfgs', multi_class='ovr').fit(data, label)
y_ = log.predict(data)
print(y_)
```

* 准确率

```python
from sklearn.metrics import accuracy_score
accuracy_score(label, y_)
```

* 分类报告

```python
from sklearn.metrics import classification_report
report = classification_report(label, y_)
print(report)
```
