# 探索式学习

## 查看对象的属性 dir()

```python
# 查看模块的属性
import random
dir(random)
```

```python
# 查看方法的属性
import random
dir(random.randint)
```

```python
# 查看类的属性
dir(list)
```

```python
# 查看对象的属性
l = list()
dir(l)
```

## 查看帮助信息 help()

```python
# 查看random模块的帮助信息
import random
help(random)
```

```python
# 查看方法randint的帮助信息
import random
help(random.randint)
```

```python
# 查看类list的帮助信息
help(list)
```

```python
# 查看对象l的帮助信息
l = list()
help(l)
```

## 查看CPython生成的字节码。分析性能时很有用

```python
import dis
help(dis.dis)
```

```python
def f1():
    l = []
    
dis.dis(f1)
```

```python
def f2():
    l = list()
    
dis.dis(f2)
```
