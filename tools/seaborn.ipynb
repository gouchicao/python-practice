# Seaborn 绘图
```
Seaborn 是基于 Python 且非常受欢迎的图形可视化库，在 Matplotlib 的基础上，进行了更高级的封装，使得作图更加方便快捷。

Seaborn是把用户常用到的可视化绘图过程进行了函数封装，形成的一个“快捷方式”，他相比Matplotlib的好处是代码更简洁，可以用一行代码实现一个清晰好看的可视化输出。而Matplotlib是可以实现高度定制化绘图的，高度定制化可以让你获得最符合心意的可视化输出结果，但也因此需要设置更多的参数，因而代码更加复杂一些。
```

## 散点图（查看两个变量之间的关系）
* [matplotlib.pyplot.scatter](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html)
* [seaborn.jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
x = np.random.randn(100)
y = np.random.randn(100)

# 用 Matplotlib 画散点图
plt.scatter(x, y)
plt.show()

# 用 Seaborn 画散点图
sns.jointplot(x, y, kind='scatter')
plt.show()
````

## 拆线图（表示数据随着时间变化的趋势）
* [matplotlib.pyplot.plot](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html)
* [seaborn.lineplot](https://seaborn.pydata.org/generated/seaborn.lineplot.html?highlight=lineplot#seaborn.lineplot)

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
x = list(range(2000, 2020))
y = np.random.randint(0, 100, len(x))

# 用 Matplotlib 画拆线图
plt.plot(x, y)
plt.show()

# 用 Seaborn 画拆线图
sns.lineplot(x, y)
plt.show()
````

## 直方图（变量的数值分布）
* [matplotlib.pyplot.hist](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html)
* [seaborn.distplot](https://seaborn.pydata.org/generated/seaborn.distplot.html?highlight=distplot#seaborn.distplot)

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
x = np.random.randint(0, 100, 100)

# 用 Matplotlib 画直方图
plt.hist(x)
plt.show()

# 用 Seaborn 画直方图
sns.distplot(x)
plt.show()
````

## 条形图（查看类别的特征）
* [matplotlib.pyplot.bar](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)
* [seaborn.barplot](https://seaborn.pydata.org/generated/seaborn.barplot.html)

````python
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
x = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
y = [5, 4, 8, 12, 7]

# 用 Matplotlib 画条形图
plt.bar(x, y)
plt.show()

# 用 Seaborn 画条形图
sns.barplot(x, y)
plt.show()
````

## 箱线图（分析出数据的差异性、离散程度和异常值等）
```
在1977年提出的，由五个数值点组成：最大值(max)，最小值(min)，中位数(median)，上下四分位数(Q3, Q1)。它可以帮我们分析出数据的差异性、离散程度和异常值等。
```
* [matplotlib.pyplot.boxplot](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.boxplot.html)
* [seaborn.boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
# 生成 0-1 之间的 10*4 维度数据
data=np.random.normal(size=(10,4)) 
lables = ['A','B','C','D']

# 用 Matplotlib 画箱线图
plt.boxplot(data, labels=lables)
plt.show()

# 用 Seaborn 画箱线图
df = pd.DataFrame(data, columns=lables)
sns.boxplot(data=df)
plt.show()
````

## 画热力图（用于多元变量分析）
```
一种矩阵表示方法，其中矩阵中的元素值用颜色来代表，不同的颜色代表不同大小的值。通过颜色就能直观地知道某个位置上数值的大小。另外你也可以将这个位置上的颜色，与数据集中的其他位置颜色进行比较。
```
* [seaborn.heatmap](http://seaborn.pydata.org/generated/seaborn.heatmap.html)

````python
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
# Seaborn 中自带的数据集 flights，记录了 1949 年到 1960 年期间，每个月的航班乘客的数量。
flights = sns.load_dataset("flights")
data=flights.pivot('year','month','passengers')

# 用 Seaborn 画热力图
sns.heatmap(data)
plt.show()
````

## 成对关系图（探索数据集中的多个成对变量的分布）
* [seaborn.pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

````python
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
"""
Seaborn 中自带的 iris 数据集，这个数据集也叫鸢尾花数据集。
鸢尾花可以分成 Setosa、Versicolour 和 Virginica 三个品种，在这个数据集中，针对每一个品种，都有 50 个数据，
每个数据中包括了 4 个属性，分别是花萼长度(sepal_length)、花萼宽度(sepal_width)、花瓣长度(petal_length)和花瓣宽度(petal_width)。
通过这些数据，需要你来预测鸢尾花卉属于三个品种中的哪一种。
"""
iris = sns.load_dataset('iris')

# 用 Seaborn 画成对关系图
sns.pairplot(iris)
plt.show()
````

## 二元关系分布（查看两个变量之间的关系）

````python
import matplotlib.pyplot as plt
import seaborn as sns

# 数据准备
# Seaborn 中自带的数据集 tips，记录了不同顾客在餐厅的消费账单及小费情况。代码中 total_bill 保存了客户的账单金额，tip 是该客户给出的小费金额。
tips = sns.load_dataset("tips")
print(tips.head())

# 用 Seaborn 画二元变量分布图（散点图，核密度图，Hexbin 图）
sns.jointplot(x="total_bill", y="tip", data=tips, kind='scatter')
sns.jointplot(x="total_bill", y="tip", data=tips, kind='kde')
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
plt.show()
````
