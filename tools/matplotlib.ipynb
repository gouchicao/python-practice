```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(0, 10)
plt.plot(n)
plt.xlabel('x number')
plt.ylabel('y number')
plt.show()
```

```python
a = np.linspace(0, 2 * np.pi, 50)
b = np.sin(a)
plt.plot(a, b)
plt.show()
```

## 绘制子图

* subplot(行数, 列数, 子图的位置）　子图的位置从1开始

```python
n1 = np.arange(0, 10)
n2 = np.arange(10, 0, -1)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(n1)
plt.subplot(1, 2, 2, facecolor='r')
plt.plot(n2)
plt.show()
```

```python
n = np.arange(1, 20)

plt.subplot(221)
plt.plot(n, np.power(n, n))
plt.title('power')

plt.subplot(222)
plt.plot(n, np.sqrt(n))
plt.title('sqrt')

plt.subplot(223)
plt.plot(n, np.log(n))
plt.title('log')

plt.subplot(224)
plt.plot(n, np.cos(n))
plt.title('cos')

plt.show()
```

## 图例

```python
x1 = np.linspace(1, 10, 10)
x2 = np.linspace(2, 15, 10)
y = np.linspace(1, 100, 10)

plt.plot(x1, y)
plt.plot(x2, y)

plt.legend(['x1', 'x2'])

plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
